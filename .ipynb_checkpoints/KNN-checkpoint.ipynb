```python
import numpy as np
from sklearn import datasets
from scipy.spatial.distance import euclidean
iris = datasets.load_iris()
y=iris.target
X=iris.data[:,:2]
```

```python
class KNearestNeighbors():
    import numpy as np
    from scipy.spatial.distance import euclidean, cityblock
    def __init__(self):
        self.x=0
        self.y=0
    
    def getDistEu(a,b):
        return euclidean(a,b)
        
        
    def getDistM(self,a,b):
        return cityblock(a,b)
        
    def fit(self, a, b):
        global X
        global y
        X=a
        y=b
        
    def distArray(self, X_test, distype='euclidean'):
        distances = []
        for i in range(len(X)):
            if distype == 'euclidean':
                print(list(X_test))
                print(list(X[i]))
                distances.append((KNearestNeighbors.getDistEu(X_test,X[i]), y[i]))
            if distype == 'manhattan':
                distances.append((getDistM(X_test,X[i]), y[i]))
        return sorted(distances, key =lambda x: x[0])
            
    def predict(self, distance = 'euclidean'):
        if distance == 'euclidean':
            getDistEu()
        elif distance == 'manhattan':
            getDistM()
        return
        
```
